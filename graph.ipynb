```python
import matplotlib.pyplot as plt

sbject=["s1","s2","s3"]
marks=[33,44,55]
color=['pink','yellow','green']
plt.figure(figsize=(5,5))
plt.pie(marks,labels=sbject,autopct='%.2f%%',explode=(0,0,0),colors=color)
plt.legend(loc=1)
plt.show()
```

```python
labels=["S1","S2","S3","S4", "S5"]
color=['orange','green','yellow','red','white']
marks=[33,30,36,23,55]
plt.figure(figsize=(10,10))


plt.subplot(2, 2,1)
plt.scatter(marks, marks, marker='*', color='black')
plt.title("Scatter Plot")

plt.subplot(2, 2, 2)
plt.bar(labels, marks, color='black')
plt.title("Bar Chart")

plt.subplot(2, 2, 3) # Bottom-left subplot
plt.pie(marks, labels=labels, autopct="%.1f%%", colors=color)
plt.title("Pie Chart")
plt.legend(loc='lower right')

plt.subplot(2, 2, 4)
plt.barh(labels, marks, color='red')
plt.title("h-Bar Chart")


plt.show()
```
